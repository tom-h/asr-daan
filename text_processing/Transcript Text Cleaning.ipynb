```python
import string
    
# return a clean text 
def clean_txt(txt):

    words = txt.split()

    word_list = []
    
    # For every word increment the frequency
    for word in words:
        # replace special characters
        word = word.replace('<slience>', '')
        
        # Remove the possible punctuation
        for char in string.punctuation:
            word = word.replace(char, '')

        # strip empty string                        
        word = word.strip()
        
        if word:
            word_list.append(word)

    return ' '.join(word_list)

def count_word(txt,frequency_dict):
    words = txt.split()
    # For every word increment the frequency
    for word in words:
        
        # strip empty string                        
        word = word.strip()
        if word:
            # Convert to lowercase                    
            word = word.lower()
            frequency_dict[word] = frequency_dict.get(word, 0) + 1

    return frequency_dict

def parse_json(jfile,freqs={}):
    with open(jfile) as json_data:
        d = json.load(json_data)
        
    n = len(d)    
    for i in range(n):
        txt = clean_txt(d[i]['transcript'])        
        d[i]['transcript'] = txt
        
        freqs = count_word(txt,freqs)
        
    return d, freqs

s = clean_txt(d[1]['transcript'])
f = count_word(s,{})
count_word(s,f)
#f1 = count_word(s,f)

jfile = 'asr-daan-codes/audio_segmenter/intermediate_example.json'
d,f = parse_json(jfile)

fn = 'asr-daan-codes/audio_segmenter/intermediate_example_normalise.json'
with open(fn,'w') as f_:
    json.dump(d,f_)
```

```python
#!/usr/bin/env python
#coding:utf-8
# Author: Alejandro Nolla - z0mbiehunt3r
# Purpose: Example for detecting language using a stopwords based approach
# Created: 15/05/13

import sys

try:
    from nltk import wordpunct_tokenize
    from nltk.corpus import stopwords
except ImportError:
    print('[!] You need to install nltk (http://nltk.org/index.html)')



#----------------------------------------------------------------------
def _calculate_languages_ratios(text):
    """
    Calculate probability of given text to be written in several languages and
    return a dictionary that looks like {'french': 2, 'spanish': 4, 'english': 0}
    
    @param text: Text whose language want to be detected
    @type text: str
    
    @return: Dictionary with languages and unique stopwords seen in analyzed text
    @rtype: dict
    """

    languages_ratios = {}

    '''
    nltk.wordpunct_tokenize() splits all punctuations into separate tokens
    
    >>> wordpunct_tokenize("That's thirty minutes away. I'll be there in ten.")
    ['That', "'", 's', 'thirty', 'minutes', 'away', '.', 'I', "'", 'll', 'be', 'there', 'in', 'ten', '.']
    '''

    tokens = wordpunct_tokenize(text)
    words = [word.lower() for word in tokens]

    # Compute per language included in nltk number of unique stopwords appearing in analyzed text
    for language in stopwords.fileids():
        stopwords_set = set(stopwords.words(language))
        words_set = set(words)
        common_elements = words_set.intersection(stopwords_set)

        languages_ratios[language] = len(common_elements) # language "score"

    return languages_ratios


#----------------------------------------------------------------------
def detect_language(text):
    """
    Calculate probability of given text to be written in several languages and
    return the highest scored.
    
    It uses a stopwords based approach, counting how many unique stopwords
    are seen in analyzed text.
    
    @param text: Text whose language want to be detected
    @type text: str
    
    @return: Most scored language guessed
    @rtype: str
    """

    ratios = _calculate_languages_ratios(text)

    most_rated_language = max(ratios, key=ratios.get)

    return most_rated_language

```

```python
text = 'Thisfdsafs fdsfsdis a fdafd fdsaf nfdaf fd afs'
language = detect_language(text)
print(language)
```
